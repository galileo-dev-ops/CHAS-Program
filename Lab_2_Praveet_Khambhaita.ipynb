## 1. Introduction to Matplotlib





Matplotlib is the most widely used plotting library for scientific computing in Python. Basically, you can think of it as Microsoft Excel's geeky brother. Most of the `matplotlib` utilities lie under the `pyplot` submodule, and are usually imported using `plt` as shorthand (similarly to `numpy` with shorthand `np`):

````python
import matplotlib.pyplot as plt
````

Furthermore, by running the following <font color="springgreen">magic command</font> (refer to section 1.4.2 of Lab 1), we will be displaying plots inline:

````python
%matplotlib inline
````

### 1.1 Plotting





The `plot()` function is used to draw points (called <font color="springgreen">markers</font>) in a diagram. By default, the function draws a line from point to point.

The function takes two parameters for specifying the points in the diagram: `plot(xs, ys)` where `xs` is an array containing the x-coordinates, and `ys` is an array containing the y-coordinates. For example:



````python
import numpy as np

xs = np.arange(0, 9, 0.1)         # Returns evenly spaced values within a given interval: np.arrange(start, stop, step_size).
ys_sin = [np.sin(x) for x in xs]  # Compute the sine function for every value x in xs.

plt.plot(xs, ys_sin)              # Plot the points using matplotlib.

plt.show()                        # Displays the plot.
````

With just a little bit of extra work we can easily plot multiple lines at once:

````python
ys_cos = [np.cos(x) for x in xs]   # Compute the cosine function for every value x in xs.

plt.plot(xs, ys_sin)               # Plot the points using matplotlib.
plt.plot(xs, ys_cos)

plt.show()                         # Displays the plot.
````

Matplotlib is remarkable because of its flexibility: you will find that there exists many ways of customising your plots.



#### 1.1.1 Markers

You can use the `marker` argument to emphasize each point with a specified marker style. For example:





````python
# Mark each point with a circle.
plt.plot(xs, ys_sin, marker = 'o')
plt.show()
````

There are many, many more marker styles out there (for e.g., stars `'*'`, crosses `'x'`, etc.). You can also change their size and colour. If you are interested, take a look at this [documentation](https://www.w3schools.com/python/matplotlib_markers.asp).

#### 1.1.2 Lines

Beyond the markers, you might want to customise the lines you plot. For instance, you can use the keyword argument `linestyle` to change the style of the plotted line:

````python
# Use a dotted line.
plt.plot(xs, ys_sin, linestyle = 'dotted')
plt.show()
````

Or use the `color` argument to change its colour:

````python
# Use a red line.
plt.plot(xs, ys_sin, color = 'red')
plt.show()
````

Or use the `linewidth` argument to change its width:

````python
# Use a thicker line.
plt.plot(xs, ys_sin, linewidth = '10')
plt.show()
````

For more examples, see this [documentation](https://www.w3schools.com/python/matplotlib_line.asp).

#### 1.1.3 Labels

We should always label our plots! For simple plots, matplotlib makes it easy: you can use the `xlabel()` and `ylabel()` functions to set a label for the x and y-axis:

````python
plt.plot(xs, ys_sin)

plt.xlabel('x axis label')         # Add a label for the x-axis.
plt.ylabel('y axis label')         # Add a label for the y-axis.

plt.show()
````

And you can use the `title()` function to set a title for the plot:

````python
plt.plot(xs, ys_sin)

plt.title('Title')       # Add a title to the plot.

plt.show()
````

You can use the `fontdict` parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels:

````python
plt.plot(xs, ys_sin)

# Define the font properties you want to use.
title_font = {
    'family':'serif',
    'color':'blue',
    'size':20
}
label_font = {
    'family':'serif',
    'color':'orange',
    'size':10
}

# Change the font properties of the title and labels.
plt.title("Title", fontdict = title_font)
plt.xlabel('x axis label', fontdict = label_font)
plt.ylabel('y axis label', fontdict = label_font)

plt.show()
````

See this [documentation](https://matplotlib.org/stable/users/explain/text/text_props.html#) for the list of properties you can include in `fontdict`.

Finally, you can also use the `legend()` function to add a key to your plot:

````python
plt.plot(xs, ys_sin)
plt.plot(xs, ys_cos)

plt.legend(['Sine', 'Cosine'])     # Add a legend to the plot: ordered list of strings.

plt.show()
````

There are still many more ways you can customise your plots to make them look the way you want them to. The best way to learn `matplotlib` is to browse through tutorials and documentations until you find something you like, copy it and then alter it to make it your own.

But to make sense of what you read, you need to know the basic plots out there: so far we have seen <font color="springgreen">line plots</font>. We move on to <font color="springgreen">scatter plots</font>.

#### 1.1.4 Scatter

With `pyplot`, you can use the `scatter()` function to draw a scatter plot. The function plots one dot for each point. Similarly to `plot()`, it requires two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:



````python
xs = [5,7,8,7,2,17,2,9,4,11,12,9,6]
ys = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(xs, ys) # Plot the points using matplotlib.

plt.show()
````

## 2. Introduction to Pandas




Pandas is a library which allows us to easily load and manipulate data from different sources: it is standardly used in AI research. It has functions for analysing, cleaning, and exploring data. `pandas` is usually imported under the `pd` shorthand (also called an <font color="springgreen">alias</font>):

````python
import pandas as pd
````

### 2.1 Series

The most fundamental data type in `pandas` is the <font color="springgreen">Series</font>. It is a one-dimensional array holding data of any type: you can think of it like the column in a table. For example:



````python
alpha = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

s = pd.Series(alpha)

print(s)
````

#### 2.1.1 Index

If nothing else is specified, the values are labeled with an <font color="springgreen">index</font> number (much like lists, but these are more explicit). The first value has index 0, second value has index 1, etc.

The index can be used to access a specified value:

````python
print(s[0]) # Returns the first element in the series.
````

However, using the `index` argument, you can name your own labels:

````python
alpha = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
index = ['first', 'second', 'third', 'fourth', 'fifth', 'sixth', 'seventh']

s = pd.Series(alpha, index=index)

print(s)
````

When you have created labels, you can access an item by referring to the label:

````python
print(s[0])         # Returns the first element in the series.
print(s['first'])   # Also returns the first element in the series.
````

For more documentation on the `Series` see [here](https://www.tutorialspoint.com/python_pandas/python_pandas_series.htm).

### 2.2 DataFrames

`Series` are not so different from lists. However, data is usually multi-dimensional (as we've seen in sections 1.2.1 and 1.2.3). Fortunately, the `pandas` library comes with a data type called a <font color="springgreen">DataFrame</font>, defined as a 2-dimensional data structure (much like a 2-dimensional array, or a table with rows and columns) of potentially different types. `Series` are like columns, a `DataFrame` is the whole table.

Taking our Ice Cream Sales example from section 1.2.1:


````python
# Put the ice cream truck data in a dictionary.
data = {
  't' : [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2],
  's' : [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408],
}

# Load the data into a DataFrame object.
df = pd.DataFrame(data)

# Display the table.
display(df)
````

As you can see from the result above, the DataFrame is like a table with rows and columns.

**Notice the presence of `index` column much like with `Series` (section 2.1.1). In fact, the `index` works in the same way for a `DataFrame`. You can choose to give you own labels or use the default one. These uniquely identify the rows of your tables.**

#### 2.2.1 Locate Rows

Now, `pandas` uses the `loc` attribute to return one or more specified row(s):

````python
print(df.loc[0]) # Returns the first row of the table.
````

````python

print(df.loc[[0, 1]]) # Returns the first and the second row of the table.
````

**Notice the that in the latter case, you need to use a list of indices.**

#### 2.2.2 Locate Columns

You can also return specific columns using their names, much as you would do with a dictionary:

````python
print(df['s']) # Returns the 's' column.
````

````python
print(df[['s', 't']]) # Returns both table columns.
````

**Much like with rows, you need to use a list of column names to return multiple columns.**

Of course, you can combine these methods to return specific items within the tables. For example:

````python
print(df['s'].loc[0])
````

Now, what else can we do with these?

#### 2.2.3 Renaming Columns

You can use the `rename` method to rename columns:

````python
# Examine the column names.
print(df.columns)
````

````python
# Rename the two columns.
df.rename(columns={'t': 'temperatures', 's': 'sales'}, inplace=True)
print(df.columns)
````

**Notice the keyword argument `inplace`. This parameter takes a boolean value, and indicates whether you wish to modify the existing `DataFrame` (`True`) or create a new one (`False`). If you don't include it, `inplace` defaults to `False`.**


See [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html) for the documentation of `rename`.

#### 2.2.4 Removing Rows/Columns


You can use the `drop` method to remove rows and columns:

````python
# Remove a single column (axis=1 refers to columns).
df.drop('sales', axis=1, inplace=True)
display(df)
````

````python
# Remove a single row using the index (axis=0 refers to rows).
df.drop(3, axis=0, inplace=True)
display(df)
````

**Note that the `axis` argument is a keyword argument which defaults to `0` (rows) if you don't provide it.**

The `inplace` argument above works in much the same way as for `rename` (section 2.2.3). Notice that is you run the cells again, you get an `KeyError`: by choosing `inplace=True`, you are modifying the original table directly.

See [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html) for the documentation.

#### 2.2.5 Insert Rows/Columns

You can add back a row using the `loc` method we encountered in section 2.2.1:

````python
# Insert the row with temperature value 15.2 back into the table.
df.loc[3] = 15.2
display(df)
````

**Notice that the order in which values are added to the `DataFrame` is maintained: the row is inserted at the end.**

You can add back a column using the `insert` method. The function takes three positional arguments `insert(loc, column, value)`:
1. `loc` is the insertion location which must be a value between `0` and `len(columns)`,
2. `column` is the label of the inserted column,
3. and `value` is an array-like object with the values of the column to be added.

````python
# Add a new sales column after the 'temperatures' column.
df.insert(1, "sales", [215, 325, 185, 406, 522, 412, 614, 544, 421, 445, 408, 332])
display(df)
````

See [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html) for the full documentation of `insert`.

#### 2.2.6 Sorting

You may find yourself in situations where you need to sort the values of one of a `DataFrame`'s columns. We do this using the `sort_values()` method. For example:

````python
out = df['sales'].sort_values() # Sort the 'sales' column in ascending order.
print(out)
print(type(out))                # Print type of this output.
````

**Notice that this sorting method returns a `Series`: it does not affect the original `DataFrame`. In other words, the sorting is temporary.**

By default, the method sort in ascending order. You can specify the ordering using the keyword argument `ascending` which takes a boolean and defaults to `True`.

````python
df['sales'].sort_values(ascending=False) # Sort the 'sales' column in descending order.
````

To perform the same operation within the `DataFrame` you can do the following:

````python
df.sort_values(['sales'], inplace=True) # Sort the DataFrame by 'sales' column in ascending order.
display(df)
````

You can also sort by multiple columns, but to make the result visible let us first add the following row to the table:

````python
df.loc[12] = [13.9, 215] # Add a row with the same sales value as row 0, but a smaller temperature.
````

````python
# Sort the DataFrame first by 'sales', then by 'temperatures' in ascending and then descending order.
df.sort_values(['sales', 'temperatures'], ascending=[True, False], inplace=True)
display(df)
````

**Notice in particular the rows `0` and `12`. The column `'sales'` is indeed sorted in ascending order, but `'temperatures'` is in descending order for equal sales values. To do this we passed a list of boolean values to the keyword `ascending` instead of just one boolean.**

#### 2.2.7 Filtering

Filtering is a very useful feature and allows you to ask specific questions of your data. For example: what were the noon temperatures of days on which the ice cream truck made a profit of £215? Let's find out:

````python
df[df['sales'] == 215] # Returns the rows of your dataset which meet the condition: sales equal to 215.
````

On such a small dataset, you might have just been able to find that out by just reading the table. But as you increase the size of your datasets and complexity of your queries, this will become much faster.

Suppose we now want to know the days on which we made a profit of £200 or more *and* where the temperature was strictly below 18°C:



````python
# The '&' operator specifies that both conditions are required: much like the 'and' boolean operation.
df[(df['sales'] >= 200) & (df['temperatures'] < 18)]
````

Similarly, you can use the `|` operator to specify that one of the two conditions are required: much like the `or` boolean operation:

```python
df[(df['sales'] >= 200) | (df['temperatures'] < 18)]
```
